```python
def configtwint(search_value):
    c = twint.Config()
    c.Since = '2021-01-01'
    c.Until = '2022-01-01'
    c.Limit = 200000
    c.Store_object = True
    c.Pandas =True
    c.Near= search_value
    c.Pandas_au = True          
    c.Pandas_clean=True
    c.Hide_output=True
    c.Count=True
    twint.run.Search(c)
```

```python
 def Run_Twint(search_vals):
    
    out_df= pd.DataFrame()
    
    for city in search_vals:
        #run twint
        configtwint(city)
        
        #get dataframe
        tweets_df = twint.storage.panda.Tweets_df
        
        #join Dataframes 
        out_df = pd.concat([out_df,tweets_df])
        
    return out_df
```

```python
#removed the irrelevant columns that came with the scraped tweets
df = df.drop(['id', 'conversation_id', 'timezone', 'place', 'cashtags', 'language', 'name', 'search', 'hashtags','user_id', 'day', 'hour', 'created_at', 'user_id_str', 'link', 'urls', 'photos', 'video',
        'thumbnail', 'retweet','nreplies', 'nretweets', 'quote_url', 'near', 'geo', 'source',  'username', 'user_rt_id', 'user_rt',
       'retweet_id', 'reply_to', 'retweet_date', 'translate', 'trans_src',
       'trans_dest','nlikes'],axis = 1)
```

```python
import re
```

```python
#used regular expression to clean the scraped data
def clean_text(text):  
    pat1 = r'@[^abc]+'                   #@signs
    pat2 = r'https?://[A-Za-z0-9./]+'  #links
    pat3 = r'\'s'                      #floating s's
    pat4 = r'\#\w+'                     # hashtags
    pat5 = r'&amp '
    pat6 = r'[^A-Za-z\s]'         #remove non-alphabet except
    combined_pat = r'|'.join((pat1, pat2,pat3,pat4,pat5, pat6))
    text = re.sub(combined_pat,"",text).lower()
    return text.strip() 
```

```python
#Applied the cleaning function on the tweets column
df["tweet"] =df["tweet"].apply(clean_text)
```

```python
#A function to calculate the polarity and subjectivity of each candidate
def getSubjectivity(text):
    return TextBlob(text).sentiment.subjectivity
def getPolarity(text):
    return TextBlob(text).sentiment.polarity

df["Subjectivity"] =df['tweet'].apply(getSubjectivity)
df["Polarity"] = df['tweet'].apply(getPolarity)
```
