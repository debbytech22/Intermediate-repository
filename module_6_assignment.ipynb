```python
import pandas as pd
import seaborn as sns
import matplotlib 
from matplotlib import pyplot as plt
import numpy as np
```

```python
car_df=pd.read_csv('Nigerian_Car_Prices.csv')
```

```python
car_df.head()
```

```python
new_columns=[]
for i in car_df.columns:
    if ' ' in i:
        colums=i.replace(' ','_')
        new_columns.append(colums)
    else:
        new_columns.append(i)
new_columns
```

```python
car_df.columns=new_columns
car_df.columns
```

```python
car_df.shape
```

```python
car_df.describe()
```

```python
car_df.isnull().sum()
```

```python
car_df['Year_of_manufacture'].fillna(0,inplace=True)
```

```python
car_df['Condition'].fillna('None',inplace=True)
```

```python
car_df['Mileage'].fillna(car_df['Mileage'].mean(),inplace=True)
```

```python
car_df.dtypes
```

```python
# car_df['Price']=car_df['Price'].replace(',','')
# car_df['Price']
new_price=[]
for i in car_df['Price']:
    new_i=int(i.replace(',',''))#.astype(float)
    new_price.append(new_i)

new_price   
```

```python
car_df['Price']= new_price
car_df['Price'].dtype
#car_df['Price'].astype(float)
```
